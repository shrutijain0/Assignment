### ***A chit fund has 25 members. Each month they contribute Rs 2000 each.  End of the month, the person who bids the lowest for the corpus, gets his bid.  The group organizer gets paid a fixed commission of Rs 2500 each month (5% of 25*2000).It is deducted from the bid winner's corpus.  The remaining amount is distributed among all the 25 customers equally.***

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Copy of Exercise.csv')
df.head()
```

```python
bidamt = np.array(df['Net amount recd by Bid winner'])
monthly_return = np.array(df['Amount returned to everyone in the group'])
total_amount=bidamt+monthly_return
```

```python
df['Total Amount']=total_amount
```

```python
df.head()
```

```python
df['Annual Return']=df['Total Amount']*(12/25)
```

```python
df.head()
```

```python
df['Return Percentage']=df['Total Amount']*(100/50000)
```

```python
df.head()
```

### ***What is the Annualized  Return of the person who bids in the last month ?***
### ***What is the Annualized Return of the person who bids in the first month ?***

```python
print('Annual  Return of the person who bids in the last month is: {}'.format(list(df['Annual Return'])[-1]))
print('Annual Return of the person who bids in the first month is: {}'.format(list(df['Annual Return'])[0]))
```
